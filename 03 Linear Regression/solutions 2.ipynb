# Exercise 2.1

## a)

```python
import numpy as np

M = np.array([
    [-0.25, -1.3],
    [-1.3, 1.25]
])

result = np.linalg.eig(M)
print(result)

lambda1, lambda2 = result.eigenvalues
v1, v2 = result.eigenvectors

V = np.array([v1, v2])
Lambda = np.array([
    [lambda1, 0],
    [0, lambda2]
])
print(V @ Lambda @ V.T)
```
